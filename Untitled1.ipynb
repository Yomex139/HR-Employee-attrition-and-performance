# importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# importing dataset

```python
df = pd.read_csv('aug_train.csv.zip')
df_copy = df.copy()
df_copy.head()
```

```python
for columns,contents in df_copy.items():
    if pd.api.types.is_numeric_dtype(contents):
        if pd.isna(contents).sum():
            df_copy[columns] = contents.fillna(contents.median())
            
for columns,contents in df_copy.items():
    if pd.api.types.is_string_dtype(contents):
        df_copy[columns] = pd.Categorical(contents).codes+1
        
```

```python
corr = df_copy.corr()
plt.figure(figsize=(20,10))
sns.heatmap(data=corr,cmap='winter',annot=True,linewidths=.5,linecolor='yellow')
```

```python
df_copy.drop(['city_development_index','company_type','company_size'],axis=1,inplace=True)
```

```python
df_copy.head()
```

```python
X = df_copy.drop('target',axis=1)
Y = df_copy['target']
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import train_test_split,cross_validate
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.linear_model import LinearRegression
RFC = RandomForestClassifier()
RFR = RandomForestRegressor()
LR = LinearRegression()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.2)
```

```python
#function to fit my model ,predict and scoring
def fit_model(model,xtrain,ytrain,xtest,ytest):
    score = {}
    model = model.fit(xtrain,ytrain)
    prediction = model.predict(xtest)
    score = accuracy_score(ytest,prediction)
    print(f'the accuracy of model,{model} is:{score*100: .2f}%')
```

```python
fit_model(model=RFC,
         xtrain=X_train,
         ytrain=Y_train,
         xtest=X_test,
         ytest=Y_test)
```

```python
cr = cross_validate(estimator=RFC,
                    X=X,
                    y=Y,
                    cv=5,
                   verbose=2,)
cr
```
